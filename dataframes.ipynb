### Copy the code from the lesson to create a dataframe full of student grades.

```python
np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)
```

```python
df
```

```python
# A. Create a column named passing_english that 
# indicates whether each student has a passing grade in english.
df['passing_english'] = df.english >= 70
```

```python
# B. Sort the english grades by the passing_english column. 
df.sort_values('passing_english')
# How are duplicates handled? 
#all duplicates are shown, sorted by passing_english, then index number
```

```python
# C. Sort the english grades first by passing_english and then by student name. 
# All the students that are failing english should be first, 
# and within the students that are failing english they should be ordered alphabetically. 
# The same should be true for the students passing english. 
# (Hint: you can pass a list to the .sort_values method)
df.sort_values('name').sort_values('passing_english')
```

```python
# D. Sort the english grades first by passing_english, and 
# then by the actual english grade, similar to how we did in the last step.
df.sort_values(['passing_english','english'])
```

```python
# E. Calculate each students overall grade and add it as a column on the dataframe. 
# The overall grade is the average of the math, english, and reading grades.
df['avg_grade'] = round((df.math + df.english + df.reading) / 3)
df
```

```python
# D. Rename the cty column to city.
mpg = mpg.rename(columns = {'cty':'city'})
```

```python
# E. Rename the hwy column to highway.
mpg = mpg.rename(columns = {'hwy':'highway'})
mpg
```

```python
# F. Do any cars have better city mileage than highway mileage?
mpg[mpg.city > mpg.highway] #no
```

```python
# G. Create a column named mileage_difference this column should 
# contain the difference between highway and city mileage for each car.
mpg['mileage_difference'] = mpg.highway - mpg.city
```

```python
# H. Which car (or cars) has the highest mileage difference?
mpg.sort_values('mileage_difference', ascending = False)
```

```python
# I. Which compact class car has the lowest highway mileage? The best?
mpg[mpg['class'] == 'compact'].sort_values('mileage_difference', ascending = False)
```

```python
# J. Create a column named average_mileage that is the mean of the city 
# and highway mileage.
mpg['average_mileage'] = round((mpg['city'] + mpg['highway']) / 2)
mpg
```

```python
# K. Which dodge car has the best average mileage? The worst?
mpg.sort_values('average_mileage', ascending=False).head(1)
```

```python
#The worst?
mpg.sort_values('average_mileage', ascending=False).tail(1)
```
