```python
titanic = pydataset.data('titanic')
```

```python
survivors = sum(titanic.survived == 'yes')
non_survivors = sum(titanic.survived == 'no')
```

```python
print(f'{survivors} survivors \n\n{non_survivors} non-survivors')
```

```python
my_labels = ['Survived', 'Did not Survive']

sizes = [survivors, non_survivors]

fig, ax = plt.subplots()
ax.pie(sizes, labels = my_labels, autopct = '%1.0f%%', colors = ('tan', 'firebrick'))

ax.set_title('Survival Rate for Titanic')
```
