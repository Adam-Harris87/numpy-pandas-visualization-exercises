## 1. Use matplotlib to plot the following equation:

y=x^2âˆ’x+2

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import random
```

```python
x = [random.randint(1,20) for _ in range(20)]
x.sort()
y = [x ** 2 - x + 2 for x in x]
print(x)
print(y)
```

```python
plt.plot(x, y)

plt.xlim(0,20.5)
plt.ylim(0,400)
plt.title('Visualization of the Equation $y=x^2âˆ’x+2$', fontsize=18)
plt.xlabel('x values')
plt.ylabel('y values')

plt.annotate('Origin', xy=(0,0), xytext=(1, 20), 
             arrowprops={'facecolor':'blue'})
plt.show;
```
