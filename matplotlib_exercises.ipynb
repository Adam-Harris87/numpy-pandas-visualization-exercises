## 1. Use matplotlib to plot the following equation:

y=x^2−x+2

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import random
import math
```

```python
x = [random.randint(1,20) for _ in range(20)]
x.sort()
y = [x ** 2 - x + 2 for x in x]
print(x)
print(y)
```

```python
plt.plot(x, y)

plt.xlim(0,20.5)
plt.ylim(0,400)
plt.title('Visualization of the Equation $y=x^2−x+2$', fontsize=18)
plt.xlabel('x values')
plt.ylabel('y values')

plt.annotate('Origin', xy=(0,0), xytext=(1, 20), 
             arrowprops={'facecolor':'blue'})
plt.show;
```

## 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

y=√x

y=x^3

y=2^x

y=1/(x+1)

You can use functions from the math module to help implement some of the equations above.

```python
x_1 = [x for x in range(21)]
x_2 = [x for x in range(11)]
x_3 = [x for x in range(11)]
x_4 = [x for x in range(11)]

x_1.sort()
x_2.sort()
x_3.sort()
x_4.sort()

y_1 = [x ** (1 / 2) for x in x_1]
y_2 = [x ** 3 for x in x_2]
y_3 = [2 ** x for x in x_3]
y_4 = [1 / (x + 1) for x in x_4]
```

```python
plt.plot(x_1, y_1)
plt.xlim(0,20.5)
plt.ylim(0,5)
plt.title('Visualization of the Equation $y=√x$', fontsize=18)
plt.xlabel('x values')
plt.ylabel('y values')

plt.show();
```

```python
plt.plot(x_2, y_2)
plt.xlim(0,10.5)
plt.ylim(0,1050)
plt.title('Visualization of the Equation $y=x^3$', fontsize=18)
plt.xlabel('x values')
plt.ylabel('y values')

plt.show();
```

```python
plt.plot(x_3, y_3)
plt.xlim(0,10.5)
plt.ylim(0,1050)
plt.title('Visualization of the Equation $y=2^x$', fontsize=18)
plt.xlabel('x values')
plt.ylabel('y values')

plt.show();
```

```python
plt.plot(x_4, y_4)
plt.xlim(0,10.5)
plt.ylim(0,1.1)
plt.title('Visualization of the Equation $y=1/(x+1)$', fontsize=18)
plt.xlabel('x values')
plt.ylabel('y values')

plt.show();
```

## 3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize=(16, 12))

plt.subplot(221)
plt.plot(x_1, y_1)
plt.xlim(0,20.5)
plt.ylim(0,5)
plt.title('$y=√x$', fontsize=14)
plt.xlabel('x values')
plt.ylabel('y values')

plt.subplot(222)
plt.plot(x_2, y_2)
plt.xlim(0,10.5)
plt.ylim(0,1050)
plt.title('$y=x^3$', fontsize=14)
plt.xlabel('x values')
plt.ylabel('y values')

plt.subplot(223)
plt.plot(x_3, y_3)
plt.xlim(0,10.5)
plt.ylim(0,1050)
plt.title('$y=2^x$', fontsize=14)
plt.xlabel('x values')
plt.ylabel('y values')

plt.subplot(224)
plt.plot(x_4, y_4)
plt.xlim(0,10.5)
plt.ylim(0,1.1)
plt.title('$y=1/(x+1)$', fontsize=14)
plt.xlabel('x values')
plt.ylabel('y values')

plt.suptitle('Visualization of Equations', fontsize = 20)

plt.show();
```

## 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.plot(x_1, y_1, label='$y=√x$')
plt.plot(x_2, y_2, label='$y=x^3$')
plt.plot(x_3, y_3, label='$y=2^x$')
plt.plot(x_4, y_4, label='$y=1/(x+1)$')

plt.xlim(0,10.5)
plt.ylim(0,20)
plt.title('Visualizations of Exponential Equations')
plt.legend()

plt.xlabel('x values')
plt.ylabel('y values')
plt.show;
```
