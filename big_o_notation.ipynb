```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import random
import math
```

```python
x1 = [x for x in range(1,11)]
y1 = [0 * x + 1 for x in x1]
x2 = [x for x in range(1,11)]
y2 = [math.log(x) for x in x2]
x3 = [x for x in range(1,11)]
y3 = [x for x in x3]
x4 = [x for x in range(1,11)]
y4 = [x* math.log(x) for x in x4]
x5 = [x for x in range(1,11)]
y5 = [x ** 2 for x in x5]
x6 = [x for x in range(1,11)]
y6 = [2 ** x for x in x6]
x7 = [x for x in range(1,11)]
y7 = [math.factorial(x) for x in x7]
x8 = [x for x in range(1,11)]
y8 = [x ** x for x in x8]
```

```python
plt.plot(x1, y1, label='O(1)')
plt.plot(x2, y2, label='O(log n)')
plt.plot(x3, y3, label='O(n)')
plt.plot(x4, y4, label='O(n log n)')
plt.plot(x5, y5, label='$O(n^2)$')
plt.plot(x6, y6, label='$O(2^n)$')
plt.plot(x7, y7, label='$O(n!)$')
plt.plot(x8, y8, label='$O(n^n)$')

plt.xlim(1,10)
plt.ylim(0,20)

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()

plt.show();
```
